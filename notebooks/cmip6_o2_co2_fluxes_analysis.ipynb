```python
units_fgco2 = 60*60*24*1000*-1*(1/44.01) ### converts to mol/m2/day
units_fgo2 = 60*60*24*-1 ### converts to mol/m2/day
seasonal_ipsl_fco2 = ipsl_fco2_SO[-120:].groupby('time.month').mean('time')*units_fgco2
seasonal_ncc_fco2 = ncc_fco2_SO[-120:].groupby('time.month').mean('time')*units_fgco2
seasonal_cccma_fco2 = cccma_fco2_SO[-120:].groupby('time.month').mean('time')*units_fgco2

seasonal_ipsl_fo2 = ipsl_fo2_SO[-120:].groupby('time.month').mean('time')*units_fgo2
seasonal_ncc_fo2 = ncc_fo2_SO[-120:].groupby('time.month').mean('time')*units_fgo2
seasonal_cccma_fo2 = cccma_fo2_SO[-120:].groupby('time.month').mean('time')*units_fgo2

seasonal_ipsl_spco2 = ipsl_spco2_SO[-120:].groupby('time.month').mean('time')
seasonal_ncc_spco2 = ncc_spco2_SO[-120:].groupby('time.month').mean('time')

seasonal_ipsl_tos = ipsl_tos_SO[-120:].groupby('time.month').mean('time')
seasonal_ncc_tos = ncc_tos_SO[-120:].groupby('time.month').mean('time')
seasonal_cccma_tos = cccma_tos_SO[-120:].groupby('time.month').mean('time')

def F_var_x_dupl(var_sc):
    ## duplicate the times axis in the seasonal cycle
    duplicated_sc = xr.concat([var_sc]*2, 'month')
    duplicated_sc['month'] = np.arange(24) + 1
    return duplicated_sc

seasonal_ipsl_fco2_dsc  = F_var_x_dupl(seasonal_ipsl_fco2)
seasonal_ncc_fco2_dsc  = F_var_x_dupl(seasonal_ncc_fco2)
seasonal_cccma_fco2_dsc  = F_var_x_dupl(seasonal_cccma_fco2)

seasonal_ipsl_fo2_dsc  = F_var_x_dupl(seasonal_ipsl_fo2)
seasonal_ncc_fo2_dsc  = F_var_x_dupl(seasonal_ncc_fo2)
seasonal_cccma_fo2_dsc  = F_var_x_dupl(seasonal_cccma_fo2)

seasonal_ipsl_tos_dsc  = F_var_x_dupl(seasonal_ipsl_tos)
seasonal_ncc_tos_dsc  = F_var_x_dupl(seasonal_ncc_tos)
seasonal_cccma_tos_dsc  = F_var_x_dupl(seasonal_cccma_tos)

seasonal_ipsl_pco2_dsc  = F_var_x_dupl(seasonal_ipsl_spco2)
seasonal_ncc_pco2_dsc  = F_var_x_dupl(seasonal_ncc_spco2)
```

```python
def F_dpCO2(pco2):
    """ compute the temperature component of the surface pCO2 rates of change"""
    
    n = np.size(pco2,0)
    
    ### rotate months to start the year in Dec
    pco2_r = np.roll(pco2,1)
    
    pco2_dt =  np.full(shape =(n), fill_value=np.nan)
    
    for i in range(24-1):
        dpco2 = pco2_r[i+1] - pco2_r[i]
        pco2_dt[i] = dpco2 

    return   pco2_dt  # pCO2_sst_sm   #pCO2_sst_sm


def F_pCO2_sst(pco2,sst):
    """ compute the temperature component of the surface pCO2 rates of change"""
    
    n = np.size(pco2,0)
    
    ### rotate months to start the year in Dec
    pco2_r = np.roll(pco2,1)
    sst_r = np.roll(sst,1)
    
    dsst_dt =  np.full(shape =(n), fill_value=np.nan)
    
    for i in range(24-1):
        dsst = sst_r[i+1] - sst_r[i]
        dsst_dt[i] = dsst

    dpCO2_sst = pco2_r[:]*0.0423*dsst_dt[:]
    
    return  dpCO2_sst  # pCO2_sst_sm   #pCO2_sst_sm


def F_pCO2_dic(pco2,sst):
    """ compute the temperature component of the surface pCO2 rates of change"""
    
    n = np.size(pco2,0)
    ### rotate months to start the year in Dec
    pco2_r = np.roll(pco2,1)
    sst_r = np.roll(sst,1)
    
    dsst_dt =  np.full(shape =(n), fill_value=np.nan)
    dpco2_dt =  np.full(shape =(n), fill_value=np.nan)
    
    for i in range(24-1):
        dsst = sst_r[i+1] - sst_r[i]
        dpco2 = pco2_r[i+1] - pco2_r[i]
        dsst_dt[i] = dsst
        dpco2_dt[i] = dpco2

    dpCO2_sst = pco2_r[:]*0.0423*dsst_dt[:]
    
    dpCO2_dic = dpco2_dt - dpCO2_sst
    
    return  dpCO2_dic
```
